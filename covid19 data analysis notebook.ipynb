## Task 2 

### 2.1: importing covid19 dataset

```python
corona_dataset_csv = pd.read_csv("covid19_Confirmed_dataset.csv")
corona_dataset_csv.head(10)
```

#### checking the shape of the dataframe

```python
corona_dataset_csv.shape
```

### 2.2: Delete the useless columns

```python
df = corona_dataset_csv.drop(["Lat", "Long"], axis = 1)
```

```python
df.head(10)
```

### 2.3: Aggregating the rows by the country

Since many rows in the column "Province/State" were missing, we grouped them by countries

```python
corona_dataset_aggregated = df.groupby("Country/Region").sum()
```

```python
corona_dataset_aggregated
```

```python
corona_dataset_aggregated.shape
```

### 2.4: Visualizing data related to a country for example China
visualization always helps for better understanding of our data.

```python
corona_dataset_aggregated.loc["China"].plot()
corona_dataset_aggregated.loc["Italy"].plot()
corona_dataset_aggregated.loc["Spain"].plot()
plt.legend()
```

### Task3: Calculating a good measure 
we need to find a good measure reperestend as a number, describing the spread of the virus in a country. 

```python
corona_dataset_aggregated.loc['China'].plot()
```

```python
corona_dataset_aggregated.loc["China"][:3].plot()
```

### 3.1: caculating the first derivative of the curve to find the rate of change

```python
corona_dataset_aggregated.loc["China"].diff().plot()
```

This graph shows the change in the infection rate day by day.

### 3.2: find maxmimum infection rate for China

```python
corona_dataset_aggregated.loc["China"].diff().max()
```

This number shows that in one day 151236 new infected cases has been recorded in China.

```python
corona_dataset_aggregated.loc["Italy"].diff().max()
```

```python
corona_dataset_aggregated.loc["Spain"].diff().max()
```

### 3.3: find maximum infection rate for all of the countries. 

```python
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []
for c in countries:
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
corona_dataset_aggregated["max_infection_rate"] = max_infection_rates
```

```python
corona_dataset_aggregated.head()
```

### 3.4: create a new dataframe with only needed column 

```python
corona_data = pd.DataFrame(corona_dataset_aggregated["max_infection_rate"])
```

```python
corona_data.head()
```

### 6.2: Deleting the useless columns

### 6.3: Aggregating the rows by the country

```python
deaths_dataset_aggregated = df2.groupby("Country/Region").sum()
```

### 6.4: Finding the maximum death rate for all of the countries. 

```python
countries = list(deaths_dataset_aggregated.index)
max_death_rate = []
for c in countries:
    max_death_rate.append(deaths_dataset_aggregated.loc[c].diff().max())
deaths_dataset_aggregated['max_death_rate'] = max_death_rate
```
